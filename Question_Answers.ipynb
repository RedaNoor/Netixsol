Lists are mutable, meaning their elements can be changed after creation. They are defined using square brackets [].

Tuples are immutable, meaning their elements cannot be changed after creation. They are defined using parentheses ().

Q5.What is a dictionary comprehension?

A dictionary comprehension is a concise way to create dictionaries. It follows a similar syntax to list comprehensions but uses curly braces {} and specifies both a key and a value.



```python
#Example:
squares = {x: x**2 for x in range(5)}
print(squares)
```

Q1. What is broadcasting in NumPy?


Broadcasting lets NumPy perform operations between arrays of different shapes without you manually looping or resizing them. NumPy automatically "stretches" the smaller array across the larger one, as long as their shapes are compatible.

Q2. How do you select all rows where a column value > 10 in a 2D array?

```python
import numpy as np

arr = np.array([[5,  12, 3],
                [11, 2,  8],
                [4,  15, 9],
                [8,  7,  6]])

result = arr[arr[:, 1] > 10]
print(result)
```

Q3. Difference between .reshape() and .flatten()?

Reshape  changes the array's dimensions (e.g. 1D → 2D) without changing the data.

Flatten always collapses to 1D, and returns a copy (safe to edit without affecting the original).

Q4. What does axis=0 vs axis=1 mean in np.sum()?


axis=0 : sum down each column (collapses rows, one result per column).

axis=1 : sum across each row (collapses columns, one result per row).
