```python
import pandas as pd
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=df)
```

#Q1. How many passengers survived?


```python
df['Survived'].value_counts()[1]
survivors= df['Survived'].value_counts()[1]
print("Total survivors:" , survivors)
```

#Q2. What was the survival rate for women vs men?


```python
women = df.loc[df.Sex == 'female']["Survived"]
rate_women = round(sum(women)/len(women) *100, 2)
print("Percentage of women who survived:", rate_women)

men = df.loc[df.Sex == 'male']["Survived"]
rate_men = round(sum(men)/len(men) *100, 2)
print("Percentage of men who survived:", rate_men)
```

#Q3. How many children survived (under 18)?


```python
children = df[df['Age'] < 18]
num_children_survived = len(children[children['Survived'] == 1])
print("Number of children who survived:", num_children_survived)
```

#Q4. What percentage of passenger were in each class?


```python
(df['Pclass'].value_counts(normalize=True) * 100).round(2)
```

#Q5. Which embarkation port had the most passengers, and what was its survival rate?

```python
print("Passenger counts by port:")
print(df['Embarked'].value_counts())

print("\nSurvival rate by port:")
print((df.groupby('Embarked')['Survived'].mean() * 100).round(2))
```

#Q6. Which class generated the most total fare revenue?


```python
df.groupby('Pclass')['Fare'].sum().round(2)
```

#Q7. What was the average fare paid by the passangers who survived vs not survived?


```python
avg_fare_survived = df[df['Survived'] == 1]['Fare'].mean().round(2)
avg_fare_not_survived = df[df['Survived'] == 0]['Fare'].mean().round(2)
overall_avg_fare = df['Fare'].mean().round(2)

print("Average fare (survived): $", avg_fare_survived)
print("Average fare (not survived): $", avg_fare_not_survived)
print("Overall average fare: $", overall_avg_fare)
```

#Q8. How many passengers paid a fare greater than $100?


```python
expensive_tickets = df[df["Fare"] > 100]

print("Passengers with Fare > $100:", len(expensive_tickets))
```
