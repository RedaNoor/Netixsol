# AFL Observation Report

## Summary
This report summarizes the findings from the cleaned AFL dataset. The observations focus on data quality, historical coverage, player statistics, team standardization, and missing biographical information. 

##  1 – Historical Data Coverage
Advanced AFL statistics are not available for the earliest seasons because they were not officially recorded. Missing values in these metrics reflect historical limitations rather than data-entry errors.

##  2 – Player Games and Scoring
Duplicate records were removed before calculating career statistics. This ensures totals such as games played, goals, and behinds accurately represent each player's recorded career.

##  3 – Team Standardization
Different spellings, capitalization, and formatting of team names were standardized into official AFL club names. This prevents duplicate categories and ensures reliable grouping and aggregation.

##  4 – Missing Biographical Information
Some players appearing in seasonal statistics do not have corresponding records in the player information dataset. These unmatched records were retained to preserve historical statistics, while missing demographic fields remain documented as a source-data limitation.
