```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df= pd.read_csv("cleaned_titanic_dataset.csv")
```

```python
sns.set_style("whitegrid")
```

**Plot a histogram of the main numeric column**

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=20, color='mediumpurple', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(['Age'])
plt.show()
```


The histogram shows the distribution of people's ages on titanic. It shows that most passengers were young adults ranging from age 15 to 35.
Very few elderly passengers (60+) were on the ship.
]

**Plot a bar chart showing average value grouped by a category column**

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Fare', data=df.groupby('Pclass')['Fare'].mean().reset_index(), palette='viridis', hue='Pclass')
plt.title('Average Fare by Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.show()
```

This bar chart compares the avergae ticket fare across different passenger classes. First class passengers paid roughly 4x more on average than secpnd class, and about 6x more than third class.

**Plot a boxplot to visualize the distribution and spot outliers visually**




```python
from matplotlib.patches import Patch

palette = {0: 'lightcoral', 1: 'seagreen'}

ax = sns.boxplot(data=df, x='Survived', y='Age', palette=palette)

ax.legend(
    handles=[
        Patch(facecolor=palette[0], label='Not Survived'),
        Patch(facecolor=palette[1], label='Survived')
    ],
    title='Survival Status'
)

plt.title('Age Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```


This boxplot helps us identify the survival rate as per age. and the points outside the box are apparently the outliers in the data.

**Plot a correlation heatmap of all numeric columns using Seaborn**



```python
numeric_cols = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation Heatmap of Numeric Columns')
plt.show()
```

The correlation heatmap displays the relationships between all the numeric columns in the dataset. The color intensity and value in each cell indicate the strength and direction of the correlation between two variables. Positive correlations are shown in warmer colors, while negative correlations are in cooler colors. Values close to 0 suggest a weak or no linear correlation.
The strongest relationship in the whole dataset is between pclass and fare.
